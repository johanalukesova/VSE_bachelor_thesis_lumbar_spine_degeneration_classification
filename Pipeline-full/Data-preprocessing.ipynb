# Theoretical part
To build a neural network for lumbar spine degenerative classification, it's essential to have a solid understanding of lumbar spine anatomy and how these structures appear on MRI.

## Lumbar spine anatomy 
The lumbar spine consists of five vertebrae (L1 to L5) located between the thoracic spine and the sacrum. 
These vertebrae are the largest and strongest in the spine, reflecting their role in supporting much of the body's weight and providing stability and flexibility.


### Lumbar vertebrae

* **Vertebral Bodies** - serve as the primary weight-bearing structures of the spine. Between each pair of vertebral bodies lies an intervertebral disc, which acts as a cushion and allows for movement between the vertebrae.
* **Intervertebral Discs** - composed of a tough outer layer called the annulus fibrosus and a gel-like core called the nucleus pulposus which together act as shock absorbers.
* **Vertebral Arch** - forms the back (posterior) part of the vertebra and surrounds the spinal canal. It consists of **pedicles**: short, thick processes that connect the vertebral body to the transverse processes and **laminae**: Flat plates of bone that extend from the pedicles and meet at the midline to form the back of the vertebral arch.
* **Spinous and Transverse Processes** - These processes serve as attachment points for muscles and ligaments and also play a role in the leverage and movement of the spine.
* **Facet Joints** - Each vertebra has four facet joints which guide and limit the range of motion of the spine.
* **Vertebral Foramina** - openings located on the sides of the vertebrae, 
* **Spinal Canal** (and Spinal Cord) - The spinal canal is the hollow passage that runs vertically through the vertebrae, formed by the alignment of the vertebral foramina. It houses and protects the spinal cord and the nerve roots as they extend down the spine.


### Vertebrae ligaments 

1. Anterior Longitudinal Ligament (ALL): Runs along the front of the vertebral bodies and discs.
2. Posterior Longitudinal Ligament (PLL): Runs along the back of the vertebral bodies inside the spinal canal.
3. Ligamentum Flavum: Connects the laminae of adjacent vertebrae and has an elastic quality.

## Spinal stenosis 
Stenosis refers to the pathological narrowing of a bodily passage or channel. In the context of the spine, particularly the lumbar region, stenosis describes the reduction in the space within the spinal canal, neural foramina, or subarticular recesses, which are essential pathways for the spinal cord and nerve roots. This narrowing can result from various degenerative changes, including hypertrophy of the facet joints, bulging of intervertebral discs, or thickening of the ligaments, such as the ligamentum flavum.

###  Types of stenosis
There are three different types of spinal stenosis: foraminal, central, and lateral. All three consist of narrowing in spine but more specifically are in different regions of your spinal canal.
1. Lateral - nerve is pinched as it exists the spinal canal.
2. Central - nerve is pinched in the main spinal canal.
3. Foraminal - nerve is pinched in the passageways located on each side of vertebrea.



## MRI imaging 
Magnetic resonance imaging (MRI) is a medical imaging technique used in radiology to form pictures of the anatomy and the physiological processes inside the body. MRI scanners use strong magnetic fields, magnetic field gradients, and radio waves to generate images of the organs in the body. MRI does not involve X-rays or the use of ionizing radiation, which distinguishes it from computed tomography (CT) and positron emission tomography (PET) scans. 

When imaging the lumbar spine, MRI is particularly valuable because it provides detailed images of both bone structures (vertebrae) and soft tissues (intervertebral discs, spinal cord, nerve roots, ligaments, and muscles). This is essential for diagnosing conditions like spinal stenosis.

### Image construction 
1. MRI machines contain a large, powerful magnet that creates a strong magnetic field around the body.
2. he human body is mostly water, and water molecules contain hydrogen atoms. When a person is placed inside the MRI machine, the magnetic field causes the protons in the hydrogen atoms to align with the direction of the magnetic field.
3. Once the protons are aligned, the MRI machine sends out radiofrequency pulses targeted at specific areas of the body. These pulses knock the protons out of alignment.
4. When the RF pulses are turned off, the protons gradually return to their original alignment with the magnetic field. As they do, they release energy in the form of radio waves, which are detected by the MRI machine's receiver coils.
5.  The MRI machine measures the energy released by the returning protons. Different tissues in the body return to alignment at different rates, depending on their composition (e.g., fat, muscle, cartilage, bone).
6.  A computer processes the detected signals and constructs them into a series of cross-sectional images (slices) of the body which can be viewed in various planes

### MRI planes 
There are three distinct orientations through whichMRI scans can be interpreted, namely sagittal, axial, and coronal.These different planes allow radiologists to examine anatomical structures from multiple perspectives, aiding in more accurate diagnosis and treatment planning.
* Axial Plane: Horizontal slices, top to bottom view.
* Sagittal Plane: Vertical slices, side to side view.
* Coronal Plane: Vertical slices, front to back view.


### Image types 
* **T1-Weighted Images:** These images provide good anatomical detail and are useful for viewing the structure of the vertebrae and the presence of fat. T1 images show fat as bright and water as dark.
* **T2-Weighted Images:** These are particularly useful for detecting pathological changes, as they highlight differences in water content. T2 images show fluid (like cerebrospinal fluid) as bright and are excellent for visualizing the intervertebral discs and identifying abnormalities such as herniations or inflammation.
* **STIR (Short Tau Inversion Recovery):** A type of T2-weighted imaging that is particularly sensitive to changes in water content, making it useful for detecting inflammation or edema in soft tissues.


### Spinal stenosis grading 
1. **Spinal Canal Stenosis:** is characterized by the narrowing of the spinal canal, which can be caused by a combination of factors, including disc bulging, ligamentum flavum thickening, and facet hypertrophy. It is best assessed on sagittal and axial T2-weighted images.
2. **Neural Foraminal Narrowing:** is typically assessed in axial and sagittal views. It appears as a reduction in the space where the nerve roots exit, often caused by disc bulges, osteophytes, or ligament thickening.

## Data preprocessing
To create a unified dataset, we need to merge these tables based on their relationships:

<div>

</div>
1. **Merge train_label with train_des:**
We use study_id and series_id as keys to bring in the series_description for each record in train_label.

2. **Merge train_label with train_data:**
For each row in train_label, we generate the appropriate severity column name by combining condition and level (lowercased, spaces replaced with underscores).
We use study_id to merge and fetch the corresponding severity from train_data.

```python
# Merge train_label with train_des on study_id and series_id
data_merged = pd.merge(train_label, train_des, on=['study_id', 'series_id'], how='left')

# Function to determine the column name in train_data
def gen_sev_col(row):
    condition_col = f"{row['condition'].lower().replace(' ', '_')}_{row['level'].lower().replace('/', '_')}"
    return condition_col

# Create a severity_col in data_merged applying gen_sev_col
data_merged['severity_col'] = data_merged.apply(gen_sev_col, axis=1)

# Merge with train_data on study_id
data_merged = pd.merge(data_merged, train_data, on='study_id', how='left')

# Extract only severity values matching with severity_col
data_merged['severity'] = data_merged.apply(lambda row: row[row['severity_col']], axis=1)

# Create unique key id
data_merged['row_id'] = (data_merged['study_id'].astype(str) + '_' + data_merged['severity_col'].astype(str))

# Filter columns
data_merged = data_merged[['row_id','study_id', 'series_id', 'condition', 'level', 'series_description', 'instance_number', 'x', 'y', 'severity']]

data_merged.head()
```

```python
# Basic function to plot the image from its path
def plot_from_path(img_path):
    ds = pydicom.dcmread(img_path) 
    plt.figure(figsize=(8, 8))
    plt.imshow(ds.pixel_array, cmap=plt.cm.gray)
    plt.title(f"Image: {img_path.split('/')[-1]}")
    plt.axis('off')
    plt.show()
```

```python
# Function to plot certain described image 
def plot_dcm_img_des(study_id, series_id, condition, level, data_merged):

    # Filter the data
    data_filter = data_merged[(data_merged['study_id'] == study_id) &
                             (data_merged['series_id'] == series_id) &
                             (data_merged['condition'] == condition) &
                             (data_merged['level'] == level)]
    
    if data_filter.empty:
        return "No matching records found."
    
    # Read and display the DICOM image
    ds = pydicom.dcmread(data_filter["img_path"].values[0])
    plt.figure(figsize=(8, 8))
    plt.axis('off')
    plt.imshow(ds.pixel_array, cmap=plt.cm.gray)
    
    # Title
    title = (f"Study ID: {study_id}, Series ID: {series_id}, Level: {level}, Instance: {data_filter['instance_number'].values[0]}\n"
             f"Condition: {data_filter['condition'].values[0]}, Severity: {data_filter['severity'].values[0]}\n"
             f"Series Description: {data_filter['series_description'].values[0]}")
    plt.title(title.strip(), fontsize=12)


    # Red circle
    for _, row in data_filter.iterrows():
        x, y = row['x'], row['y']
        plt.scatter(x, y, s=2000, edgecolors='red', facecolors='none', linewidths=3)
    
    # Display the plot
    plt.show()
```

```python
# Function to load dicom series 
def load_dic_ser(series_dir):
    dicom_files = [pydicom.dcmread(os.path.join(series_dir, f)) for f in os.listdir(series_dir) if f.endswith('.dcm')]
    dicom_files.sort(key=lambda x: int(x.InstanceNumber))
    return dicom_files

# Function to visualize dicom series 
def vis_dic_ser(dicom_series):
    for i, dicom_file in enumerate(dicom_series):
        clear_output(wait=True)
        img_shape = dicom_file.pixel_array.shape
        plt.figure(figsize=(img_shape[1]/50, img_shape[0]/50)) # Rescale the size
        plt.imshow(dicom_file.pixel_array, cmap=plt.cm.gray)
        plt.title(f'Slice {i+1} - Instance Number: {dicom_file.InstanceNumber}')
        plt.axis('off')
        plt.show()
        plt.pause(0.2)  # Adjust the speed
```

```python
def gen_records_df(dicom_dir):
    records = []
    
    for root, _, files in os.walk(dicom_dir):
        for file in files:
            if file.endswith('.dcm'):
                dicom_file_path = os.path.join(root, file)
                dicom_data = pydicom.dcmread(dicom_file_path)
                
                # Extract metadata
                study_id = dicom_data.StudyInstanceUID
                series_id_full = dicom_data.SeriesInstanceUID
                instance_number = dicom_data.InstanceNumber
                image_path = dicom_file_path
                series_id_parts = series_id_full.split('.')
                series_id = series_id_parts[1]
                
                # Append the data to the records list
                records.append({
                    "study_id": study_id,
                    "series_id": series_id,
                    "instance_number": instance_number,
                    "img_path": image_path
                })
    
    # Convert to df
    df = pd.DataFrame(records)
    return df
```

```python


# Load DICOM Image
def load_dicom_image(dicom_file_path):
    dicom_image = pydicom.dcmread(dicom_file_path)
    pixel_array = dicom_image.pixel_array
    position = np.array(dicom_image.ImagePositionPatient)
    orientation = np.array(dicom_image.ImageOrientationPatient)
    spacing = np.array(dicom_image.PixelSpacing)
    slice_thickness = dicom_image.SliceThickness if 'SliceThickness' in dicom_image else 1.0
    return pixel_array, position, orientation, spacing, slice_thickness

# Load and Resize DICOM Image
def load_and_resize_image(dicom_file_path, scale_factor=0.25):
    dicom_image = pydicom.dcmread(dicom_file_path)
    pixel_array = dicom_image.pixel_array
    resized_array = cv2.resize(pixel_array, (0, 0), fx=scale_factor, fy=scale_factor)
    position = np.array(dicom_image.ImagePositionPatient)
    orientation = np.array(dicom_image.ImageOrientationPatient)
    spacing = np.array(dicom_image.PixelSpacing) * (1 / scale_factor)
    slice_thickness = dicom_image.SliceThickness if 'SliceThickness' in dicom_image else 1.0
    return resized_array, position, orientation, spacing, slice_thickness

# Calculate Coordinates
def get_corners(position, orientation, spacing, pixel_array_shape):
    rows, cols = pixel_array_shape
    row_dir = orientation[:3]
    col_dir = orientation[3:]

    top_left = position
    top_right = position + col_dir * (cols - 1) * spacing[1]
    bottom_left = position + row_dir * (rows - 1) * spacing[0]
    bottom_right = position + row_dir * (rows - 1) * spacing[0] + col_dir * (cols - 1) * spacing[1]

    return top_left, top_right, bottom_left, bottom_right

# Create 3D Surface
def create_surface(x_coords, y_coords, z_coords, image):
    # Create a grid for interpolation
    grid_x, grid_y = np.meshgrid(np.linspace(0, 1, image.shape[1]), np.linspace(0, 1, image.shape[0]))
    
    # Interpolate x, y, and z values
    points = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
    
    grid_x_coords = griddata(points, np.array(x_coords), (grid_x, grid_y), method='linear')
    grid_y_coords = griddata(points, np.array(y_coords), (grid_x, grid_y), method='linear')
    grid_z_coords = griddata(points, np.array(z_coords), (grid_x, grid_y), method='linear')

    return grid_x_coords, grid_y_coords, grid_z_coords

# Convert Pixel to Physical Coordinates
def pixel_to_physical(x_pixel, y_pixel, position, orientation, spacing):
    x_physical = position[0] + x_pixel * orientation[0] * spacing[0] + y_pixel * orientation[3] * spacing[1]
    y_physical = position[1] + x_pixel * orientation[1] * spacing[0] + y_pixel * orientation[4] * spacing[1]
    z_physical = position[2] + x_pixel * orientation[2] * spacing[0] + y_pixel * orientation[5] * spacing[1]
    return x_physical, y_physical, z_physical


```

```python
def visualize_3d_plotly(study_id, data_merged, slice_type="both", use_surfacecolor=False, scale_factor=4):
    fig = go.Figure()
    
    # Filter data for specific study
    sample_df = data_merged[data_merged['study_id'] == study_id]
    
    # Separate sagittal and axial image paths
    sagittal_files = sample_df['img_path'][sample_df['plane'] == 'Sagittal'].unique()
    axial_files = sample_df['img_path'][sample_df['plane'] == 'Axial'].unique()
    
    # Extract conditions for sagittal images
    sagittal_conditions = [", ".join(sample_df[sample_df['img_path'] == f]['condition_lr'].unique()) for f in sagittal_files]
    
    # Extract conditions for axial images
    axial_conditions = [", ".join(sample_df[sample_df['img_path'] == f]['condition_lr'].unique()) for f in axial_files]
    
    # Define severity color mapping
    severity_colors = {
        "Normal/Mild": "green",
        "Moderate": "yellow",
        "Severe": "red"
    }
    
    # Define condition shape mapping
    condition_shapes = {
        "Spinal Canal Stenosis": "circle",
        "Neural Foraminal Narrowing": "diamond",
        "Subarticular Stenosis": "square"
    }
    
    unique_legends = set()
    
    # Load and plot the sagittal images if selected
    if slice_type in ["sagittal", "both"]:
        for sagittal_file, condition in zip(sagittal_files, sagittal_conditions):
            sagittal_image, sagittal_position, sagittal_orientation, sagittal_spacing, _ = load_and_resize_image(sagittal_file, scale_factor=1/scale_factor)
            corners = get_corners(sagittal_position, sagittal_orientation, sagittal_spacing, sagittal_image.shape)
            x_grid, y_grid, z_grid = create_surface(
                [corners[0][0], corners[1][0], corners[2][0], corners[3][0]],
                [corners[0][1], corners[1][1], corners[2][1], corners[3][1]],
                [corners[0][2], corners[1][2], corners[2][2], corners[3][2]],
                sagittal_image
            )
            sagittal_image = np.flipud(np.rot90(sagittal_image))
            
            fig.add_trace(go.Surface(x=x_grid, y=y_grid, z=z_grid, surfacecolor=sagittal_image if use_surfacecolor else None, colorscale='Gray', showscale=False))
            
            # Add severity markers with different shapes
            image_labels = sample_df[sample_df['img_path'] == sagittal_file]
            for _, row in image_labels.iterrows():
                severity_color = severity_colors.get(row['severity'], "gray")
                shape_symbol = condition_shapes.get(row['condition_lr'], "circle")
                x_pixel = row['x']//scale_factor
                y_pixel = row['y']//scale_factor
                x_physical, y_physical, z_physical = pixel_to_physical(x_pixel, y_pixel, sagittal_position, sagittal_orientation, sagittal_spacing)
                legend_key = (row['severity'], shape_symbol)
                if legend_key not in unique_legends:
                    unique_legends.add(legend_key)
                    fig.add_trace(go.Scatter3d(
                        x=[x_physical], y=[y_physical], z=[z_physical],
                        mode='markers',
                        marker=dict(symbol=shape_symbol, size=10, color=severity_color, opacity=0.8),
                        name=f"{row['severity']} ({shape_symbol})"
                    ))
    
    # Load and plot the axial images if selected
    if slice_type in ["axial", "both"]:
        for axial_file, condition in zip(axial_files, axial_conditions):
            axial_image, axial_position, axial_orientation, axial_spacing, _ = load_and_resize_image(axial_file, scale_factor=1/scale_factor)
            corners = get_corners(axial_position, axial_orientation, axial_spacing, axial_image.shape)
            x_grid, y_grid, z_grid = create_surface(
                [corners[0][0], corners[1][0], corners[2][0], corners[3][0]],
                [corners[0][1], corners[1][1], corners[2][1], corners[3][1]],
                [corners[0][2], corners[1][2], corners[2][2], corners[3][2]],
                axial_image
            )
            axial_image = np.rot90(axial_image, 1)
            
            fig.add_trace(go.Surface(x=x_grid, y=y_grid, z=z_grid, surfacecolor=axial_image if use_surfacecolor else None, colorscale='Gray', showscale=False))
            
            # Add severity markers with different shapes
            image_labels = sample_df[sample_df['img_path'] == axial_file]
            for _, row in image_labels.iterrows():
                severity_color = severity_colors.get(row['severity'], "gray")
                shape_symbol = condition_shapes.get(row['condition_lr'], "circle")
                x_pixel = row['x']//scale_factor
                y_pixel = row['y']//scale_factor
                x_physical, y_physical, z_physical = pixel_to_physical(x_pixel, y_pixel, axial_position, axial_orientation, axial_spacing)
                legend_key = (row['severity'], shape_symbol)
                if legend_key not in unique_legends:
                    unique_legends.add(legend_key)
                    fig.add_trace(go.Scatter3d(
                        x=[x_physical], y=[y_physical], z=[z_physical],
                        mode='markers',
                        marker=dict(symbol=shape_symbol, size=10, color=severity_color, opacity=0.8),
                        name=f"{row['severity']} ({shape_symbol})"
                    ))
    
    # Update layout with legend
    fig.update_layout(
        scene=dict(xaxis_title='X Axis', yaxis_title='Y Axis', zaxis_title='Z Axis'),
        legend_title_text='Severity & Condition Legend'
    )
    
    fig.show()
```

```python
# Define transformations
def dicom_to_rgb(img_path):
    """Reads a DICOM file and converts it to a 3-channel RGB image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array
    img = Image.fromarray(img_array)

    # Convert grayscale to RGB by duplicating channels
    img_rgb = img.convert("RGB")
    return img_rgb

def apply_clahe(img_path):
    """Applies CLAHE (adaptive histogram equalization) to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Convert to uint8 format
    img_uint8 = cv2.normalize(img_array, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Apply CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img_clahe = clahe.apply(img_uint8)

    # Convert back to PIL image
    return Image.fromarray(img_clahe).convert("RGB")

def apply_histogram_equalization(img_path):
    """Applies histogram equalization to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Convert to uint8 format
    img_uint8 = cv2.normalize(img_array, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Apply histogram equalization
    img_eq = cv2.equalizeHist(img_uint8)

    # Convert back to PIL image
    return Image.fromarray(img_eq).convert("RGB")

def apply_colormap(img_path, cmap=cv2.COLORMAP_JET):
    """Applies a pseudocolor map to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Normalize image to 0-255 range
    img_uint8 = cv2.normalize(img_array, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Apply the colormap
    img_colored = cv2.applyColorMap(img_uint8, cmap)

    # Convert to PIL image
    return Image.fromarray(cv2.cvtColor(img_colored, cv2.COLOR_BGR2RGB))

def apply_gamma_correction(img_path, gamma=1.5):
    """Applies gamma correction to enhance contrast in a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Normalize image
    img_norm = img_array / np.max(img_array)

    # Apply gamma correction
    img_gamma = np.power(img_norm, gamma) * 255.0
    img_gamma = np.clip(img_gamma, 0, 255).astype(np.uint8)

    # Convert to PIL image
    return Image.fromarray(img_gamma).convert("RGB")
```

```python
# Define Transformations

def apply_brightness_contrast(img_path, brightness=0.2, contrast=0.2):
    """Applies brightness and contrast adjustments to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Convert to PIL image
    img_pil = Image.fromarray(img_array).convert("RGB")

    # Apply brightness and contrast adjustments
    transform = transforms.ColorJitter(brightness=brightness, contrast=contrast)
    img_transformed = transform(img_pil)

    return img_transformed

def apply_affine_transformation(img_path, degrees=10, shear=5):
    """Applies random affine transformations (rotation, shear) to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Convert to PIL image
    img_pil = Image.fromarray(img_array).convert("RGB")

    # Apply affine transformation
    transform = transforms.RandomAffine(degrees=degrees, shear=shear)
    img_transformed = transform(img_pil)

    return img_transformed


def apply_deformation(img_path, shear_range=(0,0,0,0), degrees=0):
    """Applies vertical shear deformation to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Convert to PIL image
    img_pil = Image.fromarray(img_array).convert("RGB")

    # Ensure valid shear range (torchvision requires min < max)
    shear_range = sorted(shear_range)

    # Apply vertical deformation using affine shear
    transform = transforms.RandomAffine(degrees=degrees,  shear=shear_range)  # Only vertical shear
    img_transformed = transform(img_pil)

    return img_transformed

def apply_gaussian_noise(img_path, mean=0.3, std=0.2):
    """Applies Gaussian noise to a DICOM image."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Normalize image
    img_norm = img_array / np.max(img_array)

    # Apply Gaussian noise
    noise = np.random.normal(mean, std, img_norm.shape)
    img_noisy = img_norm + noise

    # Clip values to ensure valid image range
    img_noisy = np.clip(img_noisy * 255.0, 0, 255).astype(np.uint8)

    return Image.fromarray(img_noisy).convert("RGB")


def apply_rotation(img_path, degrees=20):
    """Applies slight rotation to a DICOM image without shearing."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Convert to PIL image
    img_pil = Image.fromarray(img_array).convert("RGB")

    # Apply rotation (without shear)
    transform = transforms.RandomAffine(degrees=degrees, shear=0)  
    img_transformed = transform(img_pil)

    return img_transformed
```

```python
def resize_image_opencv(img_path, target_size=(256, 256)):
    """Reads a DICOM image and resizes it using OpenCV."""
    ds = pydicom.dcmread(img_path)
    img_array = ds.pixel_array

    # Resize the image
    img_resized = cv2.resize(img_array, target_size, interpolation=cv2.INTER_AREA)

    return Image.fromarray(img_resized)
```
